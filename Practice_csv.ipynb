```python
from pathlib import Path
path = Path(r"D:\Programme\UBUNTU\Ubuntu.1604.2017.711.0_v1\rootfs\root\com.android.calendar\events-18-23-50\data.csv")
```

```python
csvdata = []
import csv
with open(path,"r") as fp:
    reader = csv.DictReader(fp)
    #data = next(reader)
    for row in reader: # Iterate the remaining rows
        csvdata.append(row)
csvdata
```

```python
#creates dictonary that contains the events seperated by calendar 
calendars = {} # = {calendar1Name: [Event1,...],calendar2Name: [Event1,...]}

for i in range(len(csvdata)):
    key = csvdata[i]['sync_data10'] #exchange for calendarname header from csv 
    value = csvdata[i]
    if key not in calendars:
        calendars[csvdata[i]['sync_data10']] = [value]
    elif type(calendars[key])==list:
         calendars[key].append(value)
    else:
        calendars[key] = [calendars[key][0],value]
        
```
